## 类型

按照模型学习方法进行分类

* 监督学习（Supervised Learning）

### Supervised Learning

监督学习分为三部：
1. 获取数据以及标签
2. 通过监督学习算法训练一个模型
3. 推理

![Supervised Learning](./SupervisedLearning.png)

#### 回归任务（Regression）

标签为任意数值的任务被称为回归任务，回归任务致力于回答**how many**

#### 分类任务（Classification）

标签为离散个类别的任务被称为分类任务，分类任务致力于回答**which one**

#### Tagging

一种特殊的分类任务，目标对象属于多个类，因此是多标签分类问题

#### 检索（Search）

主要与相关性度量有关，即结果不仅是一个无序的集合，更重要的是元素的顺序

#### 推荐系统（Recommender Systems）

与检索类似，推荐系统同样需要关注ranking，但是不同的是，推荐系统是一种**个性化**的服务，推荐系统的结果与用户喜好深度绑定，而不仅仅是数据间的关联性

#### 序列学习（Sequence Learning）

针对带上下文信息的数据的学习

* Tagging & Parsing：针对序列数据中特定元素的识别以及标记，元素顺序相同
* 自动音频翻译：从长序列到短序列的S2S任务，元素顺序相同
* 文本到音频：上述问题的逆任务，由短序列到长序列，元素顺序相同
* 机器翻译：变长并且顺序发生变化的S2S任务

### Unsupervised and Self-Supervised Learning

无监督学习：没有给定特定监督信号的学习任务
* 聚类：按照相似性将输入样本区分为不同的类别
* 主成分分析：描述对象的关键特征
* 因果分析
* 学习分布：典型生成式模型

自监督学习：Mask & Representation Learning

### Interacting with an Environment

上述的监督学习和无监督学习均是离线学习（offline learning），在数据采集完毕后，训练过程中不会再从环境重新获取新数据

与环境互动是一个更宏伟的目标，即模型不是单单完成预测任务，而是能够作为一个智能代理（Intelligent Agents）和环境进行交互，即模型能够影响环境


### 强化学习（Reinforcement Learning）

智能体（Agent）和环境的交互过程，旨在学习一个模型提供有效决策。

![Reinforcement Learning](./Rl.png)


环境可被完全观测：马尔可夫决策过程（Markov Decision Process）
状态不依赖于之前的操作：上下文赌博机（Contextual Bandit Problem）
没有状态仅有一组最初未知回报的可用动作：多臂赌博机（Multi-Armed Bandit Problem）
